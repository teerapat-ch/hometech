```python
import pandas as pd
df = pd.read_csv("./data/userLog_201801_201802_for_participants.csv",sep=';')
```

```python
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['year'].astype(str)+"-"+df['month'].astype(str)+"-"+df['day'].astype(str)+" "+df['hour'].astype(str)+":00:00")
df.head()
```

```python
%matplotlib inline
#user_clicked_by_hour = df.groupby(['userCode','datetime']).size().reset_index().sort_values(0,ascending=False)#[0].value_counts().median()
```

```python
#user_clicked_by_hour#[0].value_counts().median()
#df['clicked_by_hour'] =
import numpy as np
df['user_clicked_by_hour'] = df.groupby(['userCode','datetime'])['project_id'].transform(np.size)
```

```python
import numpy as np
df['user_clicked_projects_by_hour'] = df.groupby(['userCode','datetime','project_id'])['project_id'].transform(np.size)
```

```python
df.to_csv("./clean_data/userlog_feature.csv",index=False)
```
