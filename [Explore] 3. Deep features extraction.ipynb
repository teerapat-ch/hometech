```python
train_df = pd.merge(training_pair_df, train[['index','project_id','datetime']],left_on="featureIdx", right_on="index")
train_df = pd.merge(train_df,  proj_main_df[feature_col], on="project_id")
del train_df['project_id']
del train_df['index']
train_df = pd.merge(train_df, train[['index','project_id']],left_on="targetIdx", right_on="index")
train_df.columns.values[-1] = "target_project_id"
del train_df['index']

del train_df['featureIdx']
del train_df['targetIdx']
train_df.tail()
```

## Clean Data
---

Encode project_status, and do something with highest price
 

```python
# Project status
encoding_cols = ['project_status','province_id','developer_id','brand_id']
onehot_df = train_df
for encoding in encoding_cols:
    print(encoding)
    onehot_df = pd.get_dummies(train_df[encoding], prefix=encoding)
    del onehot_df["%s_%s"%(encoding,train_df[encoding].iloc[0])] # reduce dimension because not needed
    new_df = pd.concat([train_df, onehot_df],axis=1)
    del new_df[encoding]
```

```python
# Clean highest price
new_df['highest_price'] = new_df['highest_price'].fillna(method='bfill').fillna(0)
new_df['starting_price'] = new_df['starting_price'].fillna(method='bfill').fillna(0)
```

```python
# Final cleaning by removing not used columns
del new_df['developer_id']
del new_df['district_id']
del new_df['province_id']
del new_df['project_status']
del new_df['datetime']

print(new_df.columns[:20])
new_df.head()
```

## Normalize data!

```python
normalizing_cols = ["starting_price","highest_price","project_land_size_meter","lat","lon"]
for normalizing_col in normalizing_cols:
    new_df[normalizing_col] = new_df[normalizing_col]/ new_df[normalizing_col].max()
new_df.head()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.optim as optim


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()      
        self.fc1 = nn.Linear(features.shape[1], 500)
        self.fc2 = nn.Linear(500,int(target.max()))

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        
        return x


net = Net()
```

```python
torch.cuda.set_device(1)
print(torch.cuda.current_device())
```

```python
len(target_1hot)
```

```python
tensor_features = torch.Tensor(features[:20000].values).float()
tensor_targets = torch.Tensor(target[:20000].values).float()
```

```python
model = Net()

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
#values, indices = torch.max(output, 1)
```

```python
outputs = model.forward(Variable(tensor_features[:3]))
print(outputs)
loss = criterion(outputs, Variable(tensor_targets[:3].long()))
```

```python
epoch = 0
```

```python
# Training 
import math

batch_size = 1000
batch_per_epoch = math.ceil(tensor_features.shape[0]/batch_size)
epoch_num = 10

for batch_iter in range(batch_per_epoch):
    start, end = batch_iter*batch_size, (batch_iter+1)* batch_size
    X_training_batch, y_training_batch = tensor_features[start:end], tensor_targets[start:end]
    
    outputs = model(Variable(X_training_batch))
    loss = criterion(outputs, Variable(y_training_batch).long())
    
```
