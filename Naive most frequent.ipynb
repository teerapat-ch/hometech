```python
%%time
def getUserRecommend(userCode):
    district_id = list(user_mode_proj.query("userCode=='%s'"%userCode)['district_id'])
    price_range = list(user_mode_proj.query("userCode=='%s'"%userCode)['starting_price_range'])
    

    oldList = set(all_df[all_df['userCode']==userCode]['project_id'].values)

    if len(district_id)==0: 
        return "5858 3738 4201 5452 5678 5764 5067"
    else:
        possible_projs = all_df.query("district_id=='%s' & starting_price_range==%d"%(district_id[0],price_range[0]))['project_id'].value_counts()[:10+len(oldList)].index
        returning = [str(p) for p in possible_projs if p not in oldList][:7]
        if len(returning) < 7:
            possible_projs2 = all_df.query("district_id=='%s' & starting_price_range!=%d"%(district_id[0],price_range[0]))['project_id'].value_counts()[:10+len(oldList)].index
            returning2 = [str(p) for p in possible_projs2 if p not in oldList][:7]
            returning.extend(returning2)
        
        if len(returning) < 7 :
            returning.extend("5858 3738 4201 5452 5678 5764 5067".split(" "))                
        return " ".join(returning[:7])
    
users['project_id'] = users.userCode.apply(getUserRecommend)
#users.to_csv("./submission/rule_based_v3.csv",index=False)
```
